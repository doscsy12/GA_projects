## Background

In this project, data from the well known Ames housing data would be used to create regression models that will predict the sale price of houses in the area. 
<br>
<br> The Ames Housing dataset described the sale of individual residential property in Ames, Iowa from 2006 to 2010. It was found by Dr Dean De Cock, Professor of Statistics at Truman State University for his students to utilise the skills he taught in class ([source](http://jse.amstat.org/v19n3/decock.pdf)). The data set contains almost 2050 observations and a large number of explanatory variables (23 nominal, 23 ordinal, 14 discrete, and 20 continuous) involved in assessing property values. These 79-80 or so variables describe (almost) every physical attributes (quality and quantity) of the properties. Most of the variables described are features that a potential buyer would want to know before an offer was made.  


### Data

**Missing values will be examined based on data type (from the data documentation):**

* Discrete data
* Ordinal data
* Nominal data
<br>
<br> Any missing values that can be replaced will be replaced first. 
<br> Then, removal of rows still with missing data will be done, ie., remaining rows with NaN values will be dropped from the train dataset.  Special considerations (of outliers) from the data documentary will also be considered. 

```python
# find variables with missing values
pd.set_option('display.max_rows', 85)
pd.isna(train).sum()
```

How to deal with missing data:
* for continuous data, one can replace the missing values with zeros, and determine changes to mean/median/mode
* for nominal and ordinal data, one will check the data documentary if missing actually means 'none'
* if linearly related to other variables, data can be determined from those variables
* remove them, does this give a worse/better score in the model?

#### Examining ordinal variables

Since discrete and continuous data are already numerical, the ordinal variables will be examined here.
<br>
<br> Ordinal data is like categorical data but with a ranking system. Therefore, it is a form of ordered data. To examine ordinal variables, they have to be first converted to a numerical scale. 
<br> 
<br> One way to do it quickly is if the order is known. Since 'very excellent' carries the highest rank in ordinal variables such as overall_qual and overall_cond, a similar ranking will be imposed on other ordinal variables such that 'excellent' or 'good' will be given the highest rating. 
<br>


```python
train.boxplot('saleprice','garage_qual');
plt.title('Sale price grouped by quality of garage')
plt.suptitle('')
plt.xlabel('Garage quality')
plt.ylabel('Sale price'); 
```

Similarly to continuous variables, outliers are expected also in ordinal variables. One example is shown in the above plot, where there are many outliers in 'garage_qual'. While outliers, in general, are rare by definition. However, if there are many outliers, this might also imply that something interesting is going on, which may affect the relationship between the variable and the target (sale price).

```python
def find_age(row):
    '''this function calculates and returns the minimum of 2 ages'''
    age_built = row['yr_sold'] - row['year_built']
    age_remod = row['yr_sold'] - row['year_remod_add']
    row['age']=min(age_built,age_remod)
    return row
```

### Exploring ANOVA/ ChiSquare to evaluate nominal variables 

**[Disclaimer: this is experimental]**
 <br> One statistical tool explored here to evaluate whether a nominal variable affects the distribution of a continuous value, such as sale price would be the ANOVA metric. ANOVA is not the best option, but most practical in this project. So ANOVA is used to determine if any of the nominal variables are associated with sale price. To perform this, the alpha value (or level of significance) is set at 0.05. So if p-value<0.05, then the variable will be kept for the regression model. 
<br> 
<br> Chi-square test of independence is used to determine relationships between categorical variables. The null hypothesis is that there is no association between nominal variables. Similarly, the alpha value (or level of significance) is set at 0.05. So if p-value<0.05, then the variable will be kept for the regression model. 

The train_nom (dataframe with only nominal variables) will be used, together with the train['saleprice'] column.

It seems that the f_oneway anova test in scipy and anova_lm in statsmodels carry the same three assumptions:
* The samples are independent.
* Each sample is from a normally distributed population.
* The population standard deviations of the groups are all equal.


```python
# making table with multiple indices
index = pd.MultiIndex.from_product([['saleprice'],train_nom.columns],
                                   names=['cont', 'cat'])
```

```python
tables = []
for cont_var,cat_var in index:
    model = ols('{} ~ {}'.format(cont_var,cat_var), data=train).fit()
    anova_table = sm.stats.anova_lm(model, test = 'F', typ=2)

    tables.append(anova_table)

df_anova = pd.concat(tables, keys=index, axis=0)
```

```python
df_anova
```

Unfortunately, all of them showed significant differences. So either the assumptions for the anova do not hold, or something wrong was done. 
<br> 
<br> Changing the test to 'test=“Chisq”' also showed p=0.0. One post-hoc test is to perform a separate t-test or Tukey's test for each pair of groups.

### Exploring colinearity and relationships between nominal variables

Since nominal variables have no inherent ranking, finding a relationship between them are not that straight forward. But  chi-square test of independence and Cramer's V (crosstabulation table) are suitable statistical tools to assess possible relationships. 
<br>
<br> *There are, however, several assumptions before these statistics can be used:*
* there is adequate sample size
* no association can be assessed if there are missing or zero values

For the chi-squared test, the null hypothesis is that there is no significance difference in 
observed frequencies between groups are compared.

##### exploring exterior_1st and exterior_2nd

Since both are about exterior coverings on the house with exterior_2nd just the additional covering, both variables could be related.

```python
# relationship between exterior_1st and exterior_2nd
obs = pd.crosstab(train_nom['exterior_1st'],train_nom['exterior_2nd'])
chi2, p, dof, expected = stats.chi2_contingency(obs)
print(chi2,p)
```

There is a strong association between these two variables. Therefore, one of them will be dropped. 'Exterior_2nd' will be dropped as 'Exterior_1st' showed an association with 'saleprice' in the chi-square test.

<br> There are, however, other problems, namely, there are only a handful of data points per neighbourhood (see below), with some neighbourhoods having <100 data points. There could be significant collinearity between each neighbourhood, due to proxity, access to similar amenities, etc.  
<br> An alternate or associated variable would be (exact!) location or average family income of certain neighbourhoods, but these additional information are not available (on hand).

```python
train_nom['neighborhood'].value_counts()
```

Having to make an executive decision, the variable/feature to keep would be house_style. house_style seemed to be a more grouped version of ms_subclass. Even when one always hear "location, location, location!" when viewing properties, and most people seemed to know that living in the right neighbourhoods would flaunt one's style of living, and having access to niche/hip social circles, killing one's favourite children, seemed to be the norm in this project. 
<br>
<br> If necessary, ms_zoning, which identifies the zoning classification of the property can be used, since it is in the wait list.
<br>
<br> In addition, if one were to renovate the house, one would also consider the roof, since the owners would not want leaking roofs, with a roof with bad insulation (from the cold, etc). 

```python
obs = pd.crosstab(train_nom['roof_style'],train_nom['house_style'])
chi2, p, dof, expected = stats.chi2_contingency(obs)
print(chi2,p)
```

The above showed an association between roof_style and house_style. So roof_style would be dropped, since 'Gable' dominates in that variable/feature.
